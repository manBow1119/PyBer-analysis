```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set y-axis to a list of floats as the total fare in US dollars accumulated
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.40, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
#Create the plat with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label ="Boston")
#Annotate axes
plt.xlabel("Months")
plt.ylabel("Fare ($US)")
plt.ylim(0, 45)
#Create a title
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = "*", color = "red", linewidth=2, label ="Boston")
#Annotate axes
plt.xlabel("Months")
plt.ylabel("Fare ($US)")
plt.ylim(0, 45)
#Create a title
plt.title("PyBer Fare by Month")
#add grid
plt.grid()
#add legend
plt.legend()
```

```python
#Annotate using object-oriented methods
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker = "d", color = "green", linewidth = 2, label = "Boston")
ax.set_ylim(0,45)
ax.set_ylabel("Fare ($US)")
ax.set_xlabel("Month")
ax.set_title("PyBer Fare by Month")
ax.grid()
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create bar plot
plt.bar(x_axis, y_axis)
```

```python
#Annotate the plot
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#horizontal barchart
plt.barh(x_axis, y_axis)
```

```python
#Switch axes
plt.barh(y_axis, x_axis)
```

```python
#invert axes
plt.barh(x_axis, y_axis, color ="magenta", label = "Boston")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($US)")
plt.ylabel("Date")
plt.legend()
```

```python
#doing all that with OOP approach
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.plot()
```

```python
fig, ax = plt.subplots()

ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
ax.set_xlabel("Fare ($US)")
ax.set_ylabel("Month")
ax.set_title("PyBer Fare by Month")
ax.legend()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c='red', label = "Chicago")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($US)")
plt.ylabel("Date")
plt.xlim(0, 45)
plt.legend()
```

```python
#Adjust size argument to make a bubble chart
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#make data size larger for markers
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# using list comprehensions
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Bubblechart with oop
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color="skyblue", alpha =.8, edgecolor="k", linewidths = 2,label="Boston")
ax.set_xlabel("Fare ($US)")
ax.set_ylabel("Month")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.legend()
```

```python
#Creating a pie chart
#increase size of the chart

plt.subplots(figsize=(8,8))
#Explode makes a wdge pop out of the radius by a certain percentage at the index
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#autopct allows for formatting of values
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
